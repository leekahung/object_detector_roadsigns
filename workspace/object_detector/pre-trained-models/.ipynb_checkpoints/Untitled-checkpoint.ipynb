```python
# Importing Dependencies
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import xml.etree.ElementTree as ET
```

```python
# Generating path to datasets
curr_dir = os.getcwd()
main_dir = os.path.dirname(curr_dir)
datasets_dir = os.path.join(main_dir, 'datasets')

# Checking datasets directories
os.listdir(datasets_dir)
```

```python
# Defining function to load images from datasets
def load_images_from_dir(dir_path):
    images = []
    for file in sorted(os.listdir(dir_path)):
        img = cv2.imread(os.path.join(dir_path, file))
        img_res = cv2.resize(img, dsize=(256, 256))
        images.append(img)
    return images
```

```python
# Loading image files
path_to_roadsigns_dataset = os.path.join(datasets_dir, 'road_signs', 'images')

roadsigns = load_images_from_dir(path_to_roadsigns_dataset)
```

```python
# Checking images
plt.imshow(roadsigns[0])
plt.axis('off')
plt.show()
```
